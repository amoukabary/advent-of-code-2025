```python
import numpy as np

with open("day4-input.txt") as f:
    lines = f.read().splitlines()

# --- Part 1 ---
arr = np.array([[x for x in line] for line in lines])
buffer = np.zeros(arr.shape, dtype=int)
for x in range(arr.shape[0]):
    for y in range(arr.shape[1]):
        if arr[x, y] != "@":
            continue
        for dx in [0, -1, 1]:
            for dy in [0, -1, 1]:
                if dx == 0 and dy == 0:
                    continue
                nx, ny = x + dx, y + dy
                if 0 <= nx < arr.shape[0] and 0 <= ny < arr.shape[1]:
                    if arr[nx, ny] == "@":
                        buffer[x, y] += 1

count = np.sum(np.where((buffer < 4) & (arr == "@"), 1, 0))
count
```

```python
# --- Part 2 ---
arr = np.array([[x for x in line] for line in lines])
tot = 0
while True:
    buffer = np.zeros(arr.shape, dtype=int)
    for x in range(arr.shape[0]):
        for y in range(arr.shape[1]):
            if arr[x, y] != "@":
                continue
            for dx in [0, -1, 1]:
                for dy in [0, -1, 1]:
                    if dx == 0 and dy == 0:
                        continue
                    nx, ny = x + dx, y + dy
                    if 0 <= nx < arr.shape[0] and 0 <= ny < arr.shape[1]:
                        if arr[nx, ny] == "@":
                            buffer[x, y] += 1

    count = np.sum(np.where((buffer < 4) & (arr == "@"), 1, 0))
    tot += count
    arr = np.where((buffer < 4) & (arr == "@"), ".", arr)
    if count == 0:
        break
tot
```
