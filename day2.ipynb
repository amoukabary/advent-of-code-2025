```python
with open("day2-input.txt", "r") as f:
    line = f.read().strip()

ranges = line.split(",")
ranges = [((r1), (r2)) for r1, r2 in (r.split("-") for r in ranges)]
```

```python
valid_ids = []
for start, end in ranges:
    s, e = int(start), int(end)
    for i in range(s, e + 1):
        string = str(i)
        if len(string) % 2 != 0:
            continue
        if len(string) == 2:
            if string[0] == string[1]:
                valid_ids.append(i)
                continue
        if string[: len(string) // 2] == string[len(string) // 2 :]:
            # print(string)
            valid_ids.append(i)
print(sum(valid_ids))
```

```python
# Part 2
valid_ids = []
for start, end in ranges:
    s, e = int(start), int(end)
    for i in range(s, e + 1):
        string = str(i)
        for l in range(1, len(str(i))):
            if len(string) % l != 0:
                continue
            parts = [string[j : j + l] for j in range(0, len(string), l)]
            if all(part == parts[0] for part in parts):
                valid_ids.append(i)
                break


print(sum(valid_ids))
```
