```python
with open("day5-input.txt") as f:
    text = f.read()
```

```python
ranges, inputs = text.split("\n\n")
ranges = [(int(x), int(y) + 1) for x, y in [r.split("-") for r in ranges.splitlines()]]
inputs = [int(x) for x in inputs.splitlines()]
```

```python
# -- Part 1 --
fresh_stuff = []
for inp in inputs:
    fresh = False
    for x, y in ranges:
        r = range(x, y)
        if inp in r:
            fresh = True
            break
    fresh_stuff.append(fresh)
print(sum(fresh_stuff))
```

```python
# -- Part 2 --


total_fresh = 0
seen = []
sorted_ranges = sorted(ranges)
sorted_ranges
merged_ranges = []
prev_start, prev_end = sorted_ranges[0]
for r in sorted_ranges[1:]:
    start, end = r
    if start <= prev_end:
        prev_end = max(prev_end, end)
    else:
        merged_ranges.append((prev_start, prev_end))
        prev_start, prev_end = start, end
merged_ranges.append((prev_start, prev_end))

for r in merged_ranges:
    start, end = r
    total_fresh += end - start
print(total_fresh)
```
