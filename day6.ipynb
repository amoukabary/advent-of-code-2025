```python
import pprint

import numpy as np

with open("day6-input.txt") as f:
    data = f.read()
```

```python
# -- part 1 --
rows = [row.split() for row in data.splitlines()]
rows = np.array(rows)
tot = 0
for nums, operator in zip(rows[:4].T, rows[-1].T):
    nums = nums.astype(int)
    if operator == "*":
        tot += np.prod(nums)
    else:
        tot += np.sum(nums)
print(tot)
```

```python
# -- part 2 --
rows = data.splitlines()
nums = []
operator = None
tot = 0
for i in range(len(rows[0])):
    operator = operator or rows[-1][i]
    # print(operator)
    digits = [row[i] for row in rows[:-1]]
    num = "".join(digits).strip()
    if not num:
        # this means weve reached next operator
        if operator == "*":
            tot += np.prod(nums)
        else:
            tot += np.sum(nums)
        operator = None
        nums = []
    else:
        nums.append(int(num))

# final operation
if operator == "*":
    tot += np.prod(nums)
else:
    tot += np.sum(nums)
print(tot)
```
