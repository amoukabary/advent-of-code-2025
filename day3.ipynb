```python
with open("day3-input.txt") as f:
    lines = f.read().splitlines()

# part 1
results = []
for bank in lines:
    nums = [int(x) for x in bank]
    n = len(nums)
    m = max(nums[:-1])
    idx = nums.index(m)
    m2 = max(nums[idx + 1 :])
    results.append(m * 10 + m2)

print(sum(results))
```

```python
# part 2
results = []
for bank in lines:
    nums = [int(x) for x in bank]
    n = len(nums)

    # We need to select 12 digits from n digits, maintaining order
    # This is equivalent to removing (n - 12) digits to maximize the result
    k = 12  # digits to keep
    to_remove = n - k

    # Use a greedy approach: build result by always picking the largest digit
    # that leaves enough digits after it
    result = []
    start = 0

    for i in range(k):
        # We need to pick (k - i) more digits total
        # So we can look at digits from start to (n - (k - i) + 1)
        end = n - (k - i) + 1

        # Find the maximum digit in this range
        max_val = max(nums[start:end])
        max_idx = nums.index(max_val, start, end)

        result.append(max_val)
        start = max_idx + 1

    # Convert to number
    val = int(''.join(map(str, result)))
    results.append(val)

print(sum(results))
```
