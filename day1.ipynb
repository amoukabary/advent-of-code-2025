```python
INPUT_FILE = "day1-input.txt"
with open(INPUT_FILE, "r") as f:
    lines = f.readlines()
numbers = [(line.strip()) for line in lines]
rotations = [(number[0], int(number[1:])) for number in numbers]
```

```python
# Part 1
start = 50
zeros = 0
for direction, value in rotations:
    if direction == "R":
        start += value
    else:
        start -= value
    if start % 100 == 0:
        zeros += 1

print(zeros)
```

```python
# part 2
s = "0x434C49434B"
# decode hex to ascii
ascii_string = bytes.fromhex(s[2:]).decode("ascii")
print(ascii_string)
# LOL -- CLICK

pos = 50
zero_crossings = 0
for direction, value in rotations:
    full_rots, rem = divmod(value, 100)
    zero_crossings += full_rots

    rem = rem * (1 if direction == "R" else -1)

    if pos + rem >= 100 or (pos + rem <= 0 and pos != 0):
        zero_crossings += 1
    pos = (pos + rem) % 100


print(zero_crossings)
```
