```python
from pprint import pprint

with open("day7-input.txt") as f:
    data = f.read()

pprint(data)
```

```python
rows = data.splitlines()
start = rows[0].find("S")
beams = set([(start, 0)])
splits = 0
for row in range(1, len(rows)):
    new_beams = set()
    for beam in beams:
        # update beam position
        x, y = beam
        # if out of bounds, do nothing
        if x >= len(rows):
            continue
        # else, check if beam hits a splitter
        if rows[y + 1][x] == "^":
            # print("SPLIT")
            splits += 1
            new_beams.add((x - 1, y + 1))
            new_beams.add((x + 1, y + 1))
        else:
            new_beams.add((x, y + 1))
    beams = new_beams
print(splits)
```
